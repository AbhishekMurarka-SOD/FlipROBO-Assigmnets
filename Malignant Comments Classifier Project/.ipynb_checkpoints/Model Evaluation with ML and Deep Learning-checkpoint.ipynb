```python
df_class0 = data.loc[(data['malignant'] == 0) & (data['highly_malignant'] == 0) & (data['rude'] == 0) & (data['threat'] == 0) & (data['abuse'] == 0) & (data['loathe'] == 0)]
df_class1 = data[(data['malignant'] == 1) | (data['highly_malignant'] == 1) | (data['rude'] == 1) | (data['threat'] == 1) | (data['abuse'] == 1) | (data['loathe'] == 1)]
```

```python
from sklearn.utils import resample
df_class0_downsampled = resample(df_class0, 
                                 replace=False,    
                                 n_samples=16225,     
                                 random_state=42)
```

```python
df_class0_downsampled
```

```python
df=pd.concat([df_class0_downsampled,df_class1])
```

```python
## Importing more important libraries
from nltk.corpus import wordnet
import string
import nltk
import ast
import re
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.tokenize import WhitespaceTokenizer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import wordnet
from nltk.corpus import sentiwordnet
```

```python
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer() 
def clean_text(text):
    text=str(text)
    text = text.lower()
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', text)
    rem_num = re.sub('[0-9]+', '', text)
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(rem_num)  
    filtered_words = [w for w in tokens if len(w) > 2 if not w in stopwords.words('english')]
    stem_words=[stemmer.stem(w) for w in filtered_words]
    lemma_words=[lemmatizer.lemmatize(w) for w in stem_words]
    return " ".join(filtered_words)
df["comment_text"] = df["comment_text"].apply(lambda x: clean_text(x))
```

```python
y=df.drop(['id','comment_text'],axis=1)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(strip_accents='unicode', analyzer='word', ngram_range=(1,3), norm='l2')
x=vectorizer.fit_transform(df['comment_text'])
x_train,x_test,y_train,y_test = train_test_split(x,y, random_state=42, test_size=0.30, shuffle=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.multiclass import OneVsRestClassifier
```

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score
accuracy = []
precision = []
recall = []
f1_score = []
rocscore=[]
def calculate_metrics(y_test, y_pred):
    acc = metrics.accuracy_score(y_true = y_test, y_pred = y_pred)
    pre = metrics.precision_score(y_true = y_test, y_pred = y_pred,pos_label='positive',average='macro')
    rec = metrics.recall_score(y_true = y_test, y_pred = y_pred,pos_label='positive',average='macro')
    f1 = metrics.f1_score(y_true = y_test, y_pred = y_pred,pos_label='positive',average='macro')
    
    
    accuracy.append(acc)
    precision.append(pre)
    recall.append(rec)
    f1_score.append(f1)


kfold_min = []
kfold_mean = []
kfold_max = []
def calculate_kfold(estimator):
    accuracies = cross_val_score(estimator, x, y, cv = 20)
    kfold_min.append(accuracies.min())
    kfold_mean.append(accuracies.mean())
    kfold_max.append(accuracies.max())
```

```python
LogReg_pipeline = Pipeline([('clf', OneVsRestClassifier(LogisticRegression(solver='sag'), n_jobs=-1)),])
#for category in y_train:
  #  print('**Processing {} comments...**'.format(category))
LogReg_pipeline.fit(x_train, y_train)
logisticRegression_prediction = LogReg_pipeline.predict(x_test)
calculate_metrics(y_test, logisticRegression_prediction)
calculate_kfold(LogReg_pipeline)
```

```python
linsvc_pipeline = Pipeline([('lins', OneVsRestClassifier(LinearSVC(random_state=42))),])
linsvc_pipeline.fit(x_train, y_train)
linsvc_pipeline_prediction = linsvc_pipeline.predict(x_test)
calculate_metrics(y_test, linsvc_pipeline_prediction)
calculate_kfold(linsvc_pipeline)
```

```python
results = pd.DataFrame(data = {'accuracy':accuracy, 'precision': precision ,
                               'recall': recall,'f1_score': f1_score,
                               'kfold_mean': kfold_mean,'kfold_min': kfold_min,'kfold_max': kfold_max},
                      index = ['OnevsRest(LR)', 'OnevsRest(LSVC)'])
results
```
