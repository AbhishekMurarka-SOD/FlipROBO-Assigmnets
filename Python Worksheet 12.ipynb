```python
percent_missing = df.isnull().sum().round(2) * 100 / len(df)
missing_value_df = pd.DataFrame({'column_name': df.columns,
                                 'percent_missing': percent_missing})
```

```python
#From the above we can see that BuildingArea, YearBuilt, CouncilArea have missing values more than 30%
df.drop(['BuildingArea','YearBuilt','CouncilArea','Date'],axis=1, inplace= True)
```

```python
print(df.isnull().sum(axis=1))
```

```python
missing_row=df.isnull().sum(axis=1)
```

```python
percent_missing_row = df.isnull().sum(axis=1) * 100 / len(df)
missing_value_df = pd.DataFrame({'percent_missing': percent_missing_row})
```

```python
missing_value_df.sort_values('percent_missing', inplace=True)
```

```python
missing_value_df
```

```python
check_nulls = 5
if (len(df[df.isnull().sum(axis=1) > check_nulls]) / len(df)) < .2:
    filter_df = df.dropna(thresh=((df.shape[1]-check_nulls+1)), axis=0)
```

```python
df=filter_df
```

```python
df['Price'].hist()
```

```python
df['Price'].fillna(df['Price'].median(), inplace=True)
```

```python
df['Price']
```

Chose Median to fill price column as we can see from the histogram there are some outliers like some luxury houses have huge price, so taking mean is not an viable option as those outliers might affect.

```python
df.isnull().sum()
```

```python
df['Suburb'].fillna('NA',inplace=True)
df['Address'].fillna('None',inplace=True)
df['Rooms'].fillna(df['Rooms'].mean(),inplace=True)
df['Type'].fillna(df['Type'].mode()[0],inplace=True)
df['Method'].fillna(df['Method'].mode()[0],inplace=True)
df['SellerG'].fillna(df['SellerG'].mode()[0],inplace=True)
df['Distance'].fillna(df['Distance'].mean(),inplace=True)
df['Postcode'].fillna(df['Postcode'].mode()[0],inplace=True)
df['Bedroom2'].fillna(df['Bedroom2'].mode()[0],inplace=True)
df['Bathroom'].fillna(df['Bathroom'].mode()[0],inplace=True)
df['Car'].fillna(df['Car'].mode()[0],inplace=True)
df['Landsize'].fillna(df['Landsize'].mode()[0],inplace=True)
df['Lattitude'].fillna(df['Lattitude'].mode()[0],inplace=True)
df['Longtitude'].fillna(df['Longtitude'].mode()[0],inplace=True)
df['Regionname'].fillna(df['Regionname'].mode()[0],inplace=True)
df['Propertycount'].fillna(df['Propertycount'].mean(),inplace=True)
```
