Write a python program to scrap 10 images of Rayban Sunglasses from flipkart website and save them in a directory.

```python
import requests
from bs4 import BeautifulSoup
import shutil
import re
```

```python
searchs = ['rayban glass for man']
url = 'https://www.flipkart.com/search?q=rayban+glass+for+man&sid=26x&as=on&as-show=on&otracker=AS_QueryStore_HistoryAutoSuggest_1_6_na_na_na&otracker1=AS_QueryStore_HistoryAutoSuggest_1_6_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=rayban+glass+for+man%7CSunglasses&requestId=40f862ff-0070-4786-a58f-c3748624a663&as-backfill=on'
#payload = {'k': searchs}
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'}
response = requests.get(url, headers=headers)
plain_text=response.text
soup = BeautifulSoup(plain_text)
print(soup)
```

```python
count = 0
page = 1
productImgs = soup.find_all('div',{'class':'_3ZJShS _31bMyl'})
print(productImgs)
```

```python
count = 0
page = 1
while count <= 20:
    productImgs = soup.find_all('src')
    print(productImgs)
    for each in productImgs:
        img_url = each.find('src')
        response = requests.get(img_url, stream=True)
        try:
            with open('E:/Flipkart/' + count, 'wb') as out_file:
                shutil.copyfileobj(response.raw, out_file)
                count+=1
        except:
            pass
```

Write a program to scrap 20 mouses’ data which includes model name, price from www.amazon.in whose price is less than 500 and make a data frame with 2 columns “model_name”,” price” with the scraped data

```python
import pandas as pd
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'}
url='https://www.amazon.in/s?k=mouse&i=computers&rh=p_36%3A-50000&qid=1601087010&rnid=1318502031&ref=sr_nr_p_36_5'

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'lxml')
Name=[]
price=[]
count=0
while count<20:
    for i in soup.find_all('span',class_='a-size-medium a-color-base a-text-normal'):
        string=i.text
        Name.append( string.strip() )
    for i in soup.find_all('span',class_='a-price-whole'):
        price.append(i.text)
    count+=1
```

```python
df=pd.DataFrame({'Name':Name, 'Price':price})
```
